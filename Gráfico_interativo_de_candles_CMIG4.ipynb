```python
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
cmig = yf.download('CMIG4.SA',
                  start = '2021-09-08',
                  end = '2022-09-06')
```

```python
cmig['Adj Close'].plot();
```

```python
#Cria a figura
fig = make_subplots(rows = 1, cols = 1)
```

```python
fig.add_trace(go.Candlestick(name = 'Variação da cotação de CMIG4',
                             x = cmig.index, open = cmig['Open'], high = cmig['High'],
                             low = cmig['Low'], close = cmig['Close'],
                             showlegend = True), row = 1, col = 1)

#Personalizar o eixo Y
fig.update_yaxes(title_text = '<b> Preço CMIG4 (R$)', row = 1, col = 1)

#Personalizar aspectos da figura em geral, como tamanho
fig.update_layout(xaxis_rangeslider_visible = True, width = 1000, height = 500)
fig.update_layout(title = 'Gráfico interativo de candles - CMIG4')

#Mostra a figura
fig.show()
```
