```python
listings = listings_clean.sort_values('IPO_Year', ascending = True).reset_index(drop = True)
```

```python
listings.plot(subplots = True, sharex = False, figsize = (15,15))
plt.legend(fontsize = (15))
plt.show()
```

```python
listings.loc[listings.Last_Price < 200, 'Last_Price'].hist(figsize = (12,8), bins = 200)
plt.title('Last Price Frequency Distribution', fontsize = 15)
plt.xlabel('Price', fontsize = 13)
plt.ylabel('Fequency', fontsize = 13)
plt.show()
```

```python
listings.plot(kind = 'scatter',x = 'Last_Price', y = 'Market_Cap', c = 'IPO_Year', colormap = 'viridis')
plt.title('Price vs. Market Cap', fontsize = 15)
plt.xlabel('Price', fontsize = 13)
plt.ylabel('Fequency', fontsize = 13)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale = 1.1, palette = 'viridis')
sns.countplot(data = listings.loc[listings.IPO_Year > 2000],x = 'IPO_Year', hue = 'Exchange')
plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale = 1.0)
sns.barplot(data = listings,x = 'Exchange', y = 'Market_Cap', dodge = True)
plt.show()
```
