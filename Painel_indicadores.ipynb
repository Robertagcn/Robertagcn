```python
#Como montar um dashboard apenas com indicadores de rentabilidade
```

```python
#Lista de setores
fundamentus.list_papel_setor(8)
```

```python
setor = fundamentus.list_papel_setor(8)
setor
```

```python
#O query vai fazer uma consulta no seu dataframe de acordo com os criterios que voce escolher
#Trouxe apenas os dados do setor 8
df.query('index in @setor')
```

```python
df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roe', 'roic']]
```

```python
varejo = df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roe', 'roic']]
varejo
```

```python
fig = make_subplots(rows = 2,
                    cols = 2,
                    row_heights=[5,5],
                    subplot_titles=('Margem EBITDA', 'Margem Liquida', 'ROE', 'ROIC'))

fig.add_trace(go.Bar(name='Margem Ebitda', x=varejo.index, y=varejo['mrgebit']), row=1, col=1)
fig.add_trace(go.Bar(name='Margem Liquida', x=varejo.index, y=varejo['mrgliq']), row=1, col=2)
fig.add_trace(go.Bar(name='ROE', x=varejo.index, y=varejo['roe']), row=2, col=1)
fig.add_trace(go.Bar(name='ROIC', x=varejo.index, y=varejo['roic']), row=2, col=2)
fig.update_layout(title_text='<b>Overview Rentabilidade Varejo<b>', template='plotly_dark',showlegend=False)
fig.show()
```
